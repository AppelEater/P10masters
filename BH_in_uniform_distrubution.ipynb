```python
import matplotlib.pyplot as plt
import numpy as np

x = 10**(np.linspace(-5,20)/10)

fourth_the_gain = np.log2(1+x/4)

split_the_rate = np.log2(1+x)/4

plt.plot(x, fourth_the_gain, label='Fourth the gain')
plt.plot(x, split_the_rate, label='Split the rate')
plt.xlabel('SNR (dB)')
plt.ylabel('Rate (bits/s/Hz)')
plt.title('Rate vs SNR')
plt.legend()
```

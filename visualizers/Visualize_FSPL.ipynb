```python
import matplotlib.pyplot as plt
import Project_library as pl
import numpy as np
```

```python
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = [
    'CMU Serif Roman',  # primary CMU serif face
    'Times New Roman',  # fallback
    'Georgia',
    'serif'
]
plt.rcParams['axes.grid'] = True
# 3. Set the global font size:
plt.rcParams['font.size'] = 14  # change this number to taste

# 4. (Optional) Adjust axes titles and labels separately if you like:
plt.rcParams['axes.titlesize'] = 16
plt.rcParams['axes.labelsize'] = 14
```

```python
distances = np.arange(100, 1200, 1) # km
frequency = 2 # GHz
loss = []
for d in distances: 
    loss.append(pl.calculate_path_loss(d, frequency))


plt.figure(figsize=(10, 6))
plt.plot(distances, loss, label='Path Loss', color='blue')
plt.xlabel('Distance (km)')
plt.ylabel('Path Loss (dB)')
plt.savefig("Path_loss_vs_distance.pdf")
```
